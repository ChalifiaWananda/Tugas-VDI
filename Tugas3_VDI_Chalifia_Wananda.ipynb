1. Data yang digunakan pada visualisasi kali ini adalah data diabetes.
2. Tujuan Visualisasi : Masyarakat.

  Fungsi Visualisasi : untuk melihat hubungan antara prediksi masyarakat yang menderita diabates dengan kadar glukosa yang dimilikinya.

  Tone : -

3. Visualisasi yang digubakan adalah scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("/content/diabetes.csv")

plt.figure(figsize=(10, 6))
plt.scatter(df['DiabetesPedigreeFunction'], df['Glucose'], color='green', marker='o')

plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('Glucose')
plt.title('Scatter Plot using Matplotlib')
plt.grid(True)
plt.show()
```
